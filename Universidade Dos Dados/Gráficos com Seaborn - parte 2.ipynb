```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.barplot(x='Species', y='SepalLengthCm', data=iris);
```

```python
sns.barplot(y='Species', x='SepalLengthCm', data=iris);
```

```python
iris['Agrupamento_Petala'] = np.where(iris['PetalLengthCm'] >= 6, 'Grande', 'Pequena')
iris.head()
```

```python
sns.barplot(x='Species', y='SepalLengthCm', hue='Agrupamento_Petala',data=iris, errorbar=('ci', 20));
```

```python
sns.barplot(x='Species', y='SepalLengthCm', hue='Agrupamento_Petala',data=iris, palette='Paired');
# veja mais sobre paletas:
# https://seaborn.pydata.org/tutorial/color_palettes.html
```
