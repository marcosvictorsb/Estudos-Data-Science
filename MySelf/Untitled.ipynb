```python
import pandas as pd
import seaborn as sns
```

```python
salarios = pd.read_csv('ds_salaries.csv')
```

```python
salarios.head()
```

```python
media_salario_anual = salarios.groupby(['work_year', 'experience_level'])['salary'].mean().reset_index()
media_salario_anual
```

```python
# media salarial por ano baseado em sua experiÃªncia.
sns.barplot(data=media_salario_anual, x='work_year', y='salary', hue='experience_level');
```
